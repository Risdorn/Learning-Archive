## 2. Create an instance of a PydanticAI agent

````python
!pip install -q pydantic_ai openai
````

Here, we are creating a custom OpenAI compatible endpoint for the agent

````python
from pydantic_ai.models.openai import OpenAIModel
from pydantic_ai.providers.openai import OpenAIProvider

model_id = "Qwen/Qwen3-1.7B"
openai_api_key = "EMPTY"
openai_api_base = "http://localhost:8000/v1"

provider = OpenAIProvider(
    api_key=openai_api_key,
    base_url=openai_api_base,
)
agent_model = OpenAIModel(model_id, provider=provider)
````

Using the endpoint, we create an instance of the Agent class

````python
from pydantic_ai import Agent

agent = Agent(model=agent_model)
````

To test our agent, we define a helper function.

You can refer [pydantic_ai](https://ai.pydantic.dev/agents/#running-agents) for other ways to interact with the agent

````python
import asyncio
from pydantic_ai.mcp import MCPServerStdio

async def run_async(prompt: str) -> str:
    async with agent.run_mcp_servers():
        result = await agent.run(prompt)
        return result.output
````

````python
await run_async("What is the capital of India?")
````

````python
await run_async("What is Agentic AI")
````

## 3. Create a date/time tool for the agent

````python
await run_async("WhatвЂ™s the date today?")
````

We can make the model run custom functions to get the required information using "Tool/Function Calling". Here we will use the `Tool` decorator from `pydantic_ai` for it

````python
from datetime import datetime
from pydantic_ai import Tool
@Tool
def get_current_date() -> str:
    """Return the current date/time as an ISO-formatted string."""
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")
````

We now need to inform the model about this tool and when to use it. For that we create another instance of the Agent class with this information

````python
agent = Agent(
    model=agent_model,
    tools=[get_current_date],
    system_prompt = (
        "You have access to:\n"
        "   1. get_current_time(params: dict)\n"
        "Use this tool for date/time questions."
    )
)
````

````python
await run_async("WhatвЂ™s the date today?")
````

## 4. Use an MCP server for the date/time tool

[MCP](https://modelcontextprotocol.io/docs/getting-started/intro) servers provide

    вњ… Standardized API interfaces

    рџ”„ Reusability across projects

    рџ“¦ Prebuilt functionality


````python
!pip install mcp-server-time
````

````python
from pydantic_ai.mcp import MCPServerStdio

time_server = MCPServerStdio(
    "python",
    args=[
        "-m", "mcp_server_time",
        "--local-timezone=Asia/Kolkata",
    ],
)
````

````python
agent = Agent(
    model=agent_model,
    toolsets=[time_server],
    system_prompt = (
        "You are a helpful agent and you have access to this tool:\n"
        "   get_current_time(params: dict)\n"
        "When the user asks for the current date or time, call get_current_time.\n"
    )
)
````

````python
await run_async("WhatвЂ™s the date today?")
````

## 5. Turn the agent into an Airbnb finder

````python
airbnb_server = MCPServerStdio(
    "npx", args=["-y", "@openbnb/mcp-server-airbnb", "--ignore-robots-txt"]
)
````

````python
system_prompt = """
You have access to three tools:
1. get_current_time(params: dict)
2. airbnb_search(params: dict)
3. airbnb_listing_details(params: dict)
When the user asks for listings, first call get_current_time, then airbnb_search, etc.
"""

agent = Agent(
    model=agent_model,
    toolsets=[time_server, airbnb_server],
    system_prompt=system_prompt,
)
````

````python
await run_async("Find a place to stay in Jammu for next Sunday for 3 nights for 2 adults?")
````
